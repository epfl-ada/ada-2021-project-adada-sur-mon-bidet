## Vocabulary construction

trusted sources > synonym generation > hand filtering

```python
# https://www.vocabulary.com/lists/1491890
base = "aerosol agriculture atmosphere agriculture atmosphere biosphere carbon climate climatology coral desertification earth emission fossil fuel geosphere glacier warming greenhouse hydrosphere methane precipitation solar temperature"
# https://en.wikipedia.org/wiki/Climate_change
base += "emissions energy gas gases petrol wildfires permafrost extinct extinction IPCC COP coal nuclear GHG drought hurricane typhoon net-zero deforestation sustainable sustainability hydro"
#https://19january2017snapshot.epa.gov/climatechange/glossary-climate-change-terms_.html
base += "albedo anthropogenic atmosphere biofuels biomass footprint sequestration cryosphere ecosystem biodiversity ozone recycling resilience wastewater"
#https://www.health.state.mn.us/communities/environment/climate/docs/film/vocab_list.pdf
base += "mitigation anthropocene AR4 AR5 CO2 "

base = base.split()
len(base)
```

```python
synonyms = []

for w in base:
  for syn in wordnet.synsets(w):
    for l in syn.lemmas():
      synonyms.append(l.name())

climate_dict = set([*base,*synonyms])
print(len(climate_dict))
```

```python
print(climate_dict)
climate_dict = climate_dict - set(["C", "husbandry", "vim", "nail", "fuzz", "ambience", "muscularity", "mood", "gas_pedal", "arrest", "bull", "blow", "push", "brag", "snitch", "pick-up", "step", "zip", "pig", "vigor", "nursery", "atm", "thieve", "spray_can", "char"])
climate_dict = set([w.replace("_", " ") for w in climate_dict])
print(len(climate_dict))
print(climate_dict)
```

## IO

```python
with open('base_climate_dictionary.txt', 'w') as f:
    for word in base:
        f.write("%s\n" % word)

!wc base_climate_dictionary.txt

with open('wide_climate_dictionary.txt', 'w') as f:
    for word in climate_dict:
        f.write("%s\n" % word)

!wc wide_climate_dictionary.txt
```
